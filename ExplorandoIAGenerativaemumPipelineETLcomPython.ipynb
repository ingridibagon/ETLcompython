# Explorando IA Generativa em um Pipeline de ETL com Python

**Contexto:** Neste projeto vou salvar a informaçao da página de livros mais vendidos da Amazom.com.br e extraer a informaçao e salvar em um arquivo .csv. E depos usasr a API do ChatGPT para gerar uma mensagem de marketing para cada livro.

**Etapas do Projeto:**

1. Baixar o conteudo da página usando 'requests'
  
2. Extrair  nome do livro e o nome do autor.
3. Compilar a informaçao e salvá-la em um arquivo .csv
3. Depois, usar a API do ChatGPT (OpenAI) para gerar uma mensagem de marketing personalizada para cada livro.
4. Uma vez que a mensagem para cada livro esteja pronta, será adicionada em uma nova columna do arquivo .csv



## **E**xtract

1. Baixar a lista de mais vendidos usando 'requests'

```python
import requests
url="https://www.amazon.com.br/gp/bestsellers/books/?ie=UTF8&ref_=sv_b_2"


def get_url(url):
  response = requests.get(url)
  return response.text if response.status_code == 200 else None

response=get_url(url)
print(response)
```

Uso BeatifulSoup para converter request.text em htl e os metodos do objeto doc permitem procurar a informaçao que estou querendo dos livros

```python
from bs4 import BeautifulSoup
doc=BeautifulSoup(response, 'html.parser')
```

```python
# extraindo o nome do livro
book_title_tags= doc.find_all('div',{"class": "zg-grid-general-faceout"})
# book_title_tags[0].find('a',{'class':'a-link-normal'})
book_title_tags[0].find('span').text
```

```python
#funçao para extrair o nome dos livros
def get_nome_livro(doc):
    book_title_tags= doc.find_all('div',{"class": "zg-grid-general-faceout"})
    #lista par adicionar os nomes dos livros mais vendidos
    livros=[]
    for tag in book_title_tags:
        nome_livro = tag.find('span')
        livros.append(nome_livro.text)
    return livros
get_nome_livro(doc)[:20] #Top 20 livros mais vendidos Amazon.com.br
```

```python
#funçao para extrair o nome do autor
def get_autores(doc):
    autores= doc.find_all('div',{'class':'zg-grid-general-faceout'})
    nome_autores=[]
    for tag in autores:
        nome_autores.append(tag.find('div',{'class':'a-row a-size-small'}).text)

    return nome_autores
get_autores(doc)[0]
```

```python
#Criando um diccionario com o nome do livro e o nome do autor
Top_10={'Título': [], 'Autor': []}

Top_10['Título'] = get_nome_livro(doc)[:10]
Top_10['Autor'] = get_autores(doc)[:10]
print(Top_10)
```

```python
import pandas as pd
df_Top_10 = pd.DataFrame.from_dict(Top_10, orient= 'index')
df_Top_10= df_Top_10.transpose()
df_Top_10
```

```python
import openai

openai.api_key = openai_api_key

def generate_ai_propaganda(n_livro,Top_10):
  completion = openai.ChatCompletion.create(
    model="gpt-4",
    messages=[
      {
          "role": "system",
          "content": "Você é um especialista em marketing digital."
      },
      {
          "role": "user",
          "content": f"Crie uma mensagem para vender o livro {Top_10['Título'][n_livro]} do autor {Top_10['Autor'][n_livro]} (máximo de 200 caracteres)"
      }
    ]
  )
  return completion.choices[0].message.content.strip('\"')
marketing=[]
for n in range(10):
  propaganda = generate_ai_propaganda(n,Top_10)
  marketing.append(propaganda)
  print(propaganda)
```
